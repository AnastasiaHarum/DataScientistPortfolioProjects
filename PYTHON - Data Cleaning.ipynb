# Mengatasi Missing Value

### Mengisi Nilai Kosong Pada Data Menggunakan Mean

```python
#Mengisi nilai kosong pada variabel CO(GT)
mean_value = data['CO(GT)'].mean()
data['CO(GT)'] = data['CO(GT)'].fillna(mean_value)

#Mengisi nilai kosong pada variabel PT08.S1(CO)
mean_value = data['PT08.S1(CO)'].mean()
data['PT08.S1(CO)'] = data['PT08.S1(CO)'].fillna(mean_value)

#Mengisi nilai kosong pada variabel C6H6(GT)
mean_value = data['C6H6(GT)'].mean()
data['C6H6(GT)'] = data['C6H6(GT)'].fillna(mean_value)

#Mengisi nilai kosong pada variabel PT08.S2(NMHC)
mean_value = data['PT08.S2(NMHC)'].mean()
data['PT08.S2(NMHC)'] = data['PT08.S2(NMHC)'].fillna(mean_value)

#Mengisi nilai kosong pada variabel NOx(GT)
mean_value = data['NOx(GT)'].mean()
data['NOx(GT)'] = data['NOx(GT)'].fillna(mean_value)

#Mengisi nilai kosong pada variabel PT08.S3(NOx)
mean_value = data['PT08.S3(NOx)'].mean()
data['PT08.S3(NOx)'] = data['PT08.S3(NOx)'].fillna(mean_value)

#Mengisi nilai kosong pada variabel NO2(GT)
mean_value = data['NO2(GT)'].mean()
data['NO2(GT)'] = data['NO2(GT)'].fillna(mean_value)

#Mengisi nilai kosong pada variabel PT08.S4(NO2)
mean_value = data['PT08.S4(NO2)'].mean()
data['PT08.S4(NO2)'] = data['PT08.S4(NO2)'].fillna(mean_value)

#Mengisi nilai kosong pada variabel PT08.S5(O3)
mean_value = data['PT08.S5(O3)'].mean()
data['PT08.S5(O3)'] = data['PT08.S5(O3)'].fillna(mean_value)

#Mengisi nilai kosong pada variabel T
mean_value = data['T'].mean()
data['T'] = data['T'].fillna(mean_value)

#Mengisi nilai kosong pada variabel RH
mean_value = data['RH'].mean()
data['RH'] = data['RH'].fillna(mean_value)

#Mengisi nilai kosong pada variabel AH
mean_value = data['AH'].mean()
data['AH'] = data['AH'].fillna(mean_value)
```

```python
data
```

```python
data.isnull().sum()
```

# Menghilangkan Outlier

```python
data
```

```python
q1 = data[['CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH']].quantile(0.25)
q1
```

```python
q3 = data[['CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH']].quantile(0.75)
q3
```

```python
IQR = q3 - q1
IQR
```

```python
lower_limit = q1 - (1.5 * IQR)
lower_limit
```

```python
upper_limit = q3 + (1.5 * IQR)
upper_limit
```

```python
df_data = pd.DataFrame(np.where(data[['CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH']] < lower_limit, lower_limit,
         (np.where(data[['CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH']] > upper_limit, upper_limit, data[['CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH']]))), columns=[['CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH']])
df_data
```

### Pengetahuan yang diperoleh dari dataset Air Quality  UCI adalah  Dataset tersebut berisikan tentang faktor-faktor yang mempengaruhi kualitas udara, dimana data ini direkam selama 1 tahun mulai dari Maret 2006 hingga Februari 2005 dimana dataset  ini berisi 9358 sampel respon rata-rata perjam dari 5 sensor kimia oksida logam yang tertanam dalam Perangkat Multisensor Kimia Kualitas Udara, perangkat ini terletak di area yang sangat tercemar, permukaan jalan, di dalam kota Italia. Data ini berisi 9356 baris dan 15 kolom atau variabel yaitu Date, Time, CO(GT), PT08.S1(CO),	NMHC(GT), C6H6(GT),	PT08.S2(NMHC), NOx(GT), PT08.S3(NOx), NO2(GT), PT08.S4(NO2), PT08.S5(O3), T, RH, dan AH. Setelah dilakukan noisy pada data menggunakan python didapatkan informasi bahwa banyak data yang hilang, terutama di variabel NMHC(GT) yang berisi 8443 data Nan sehingga kami memutuskan untuk tidak menggunakan variabel NMHC(GT) karena hampir sebagian besar data hilang. Namun, untuk variabel Date dan Time tidak memiliki nilai Nan. Kemudian dilakukan pengisian nilai terhadap variabel yang berisi nilai Nan dengan menggunakan nilai mean dari masing-masing variabel. Selanjutnya dilakukan deteksi outlier pada masing-masing variabel yang dimana hampir semua variabel memiliki outlier kecuali variabel RH, jadi perlu dilakukan penghapusan terhadap outlier dengan menggunakan pembatasan batas bawah 25% dan batas atas 75% setelah itu dapat dilakukan normalisasi data dengan cara membuat rentang nilai yang sama yaitu dengan batas dari 0 sampai 1 maka tidak ada nilai yang terlalu besar dan terlalu kecil sehingga menghasilkan data akhir yang berdistribusi normal. Berdasarkan informasi dari data dan proses-proses data mining seperti mengatasi noisy, missing value, deteksi outlier, dan normalisasi, kita dapat menyimpulkan bahwa data Air Quality UCI sangat penting dan berguna karena berisikan banyak informasi mengenai faktor-faktor yang mempengaruhi kualitas udara sehingga dataset ini bisa digunakan untuk menilai pencemaran udara di suatu daerah.
